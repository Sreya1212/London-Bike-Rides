```python
#renaming the column names:
new_cols={'timestamp':'time',
         'cnt':'count',
         't1':'temp_real_C',
         't2':'temp_feels_like_C',
         'hum':'humidity_precent',
         'wind_speed':'wind_speed_kph',
         'weather_code':'weather',
         'is_holiday':'is_holiday',
         'is_weekend':'is_weekend',
         'season':'season'}
df.rename(new_cols,axis=1,inplace=True)

#changing humidity values to percentage
df['humidity_percent']=df.humidity_precent/100
df
```

```python
#creating season dictionary to map integers 0-3 to each actual values
season_dict={'0.0':'spring',
            '1.0':'summer',
            '2.0':'autumn',
            '3.0':'winter'}

#creating weather dictionary to map integers to each actual values
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}
```

```python
#changing column type to string
df.season=df.season.astype('str')
df.weather=df.weather.astype('str')
```

```python
#mapping new dictionary
df.season=df.season.map(season_dict)
df.weather=df.weather.map(weather_dict)
```
